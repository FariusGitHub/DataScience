```python
!pip install kora -q
from kora.selenium import wd
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from tqdm import tqdm 

wd.get("https://3dexperience.virtualtester.com/")

time.sleep(2)
ht_pref = '/html/body/div[1]/div/div[2]/div/div/div'
element = wd.find_element(By.XPATH, ht_pref+'[1]/ul/li[5]')
wd.execute_script("arguments[0].click();", element)
time.sleep(1)

prefix = ht_pref + \
  '[2]/div[2]/div/div/div/div/div[1]/div/div/div[1]/div/div/div/div/div'

wd.find_element(By.XPATH, prefix + '[2]/div/div[2]/div[1]/input').\
  send_keys('Canada', Keys.ENTER)

for i in tqdm(range(4)):
  wd.find_element(By.XPATH, prefix + '[1]/div/div[2]/div[1]/input').click()
  wd.find_element(By.XPATH, prefix + '[1]/div/div[2]/div[1]/input').\
    send_keys(Keys.ARROW_DOWN)
  wd.find_element(By.XPATH, prefix + '[1]/div/div[2]/div[1]/input').\
    send_keys(Keys.ENTER)
  time.sleep(2)

url2 = prefix[:70]+'[2]/div/div/div/div/div[1]/div[2]/div[2]/div[1]'

wd.find_element(By.XPATH, url2).click()

for i in tqdm(range(1483)):
  wd.find_element(By.XPATH, url2).send_keys(Keys.DOWN)

time.sleep(2)
display(wd)
```

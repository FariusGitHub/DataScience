```
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from numpy import genfromtxt

data = genfromtxt('https://farius.s3.amazonaws.com/mortgage.csv', delimiter=',')

x_labels=['Jan2018\n(01)','Feb2018\n(02)','Mar2018\n(03)','Apr2018\n(04)',
          'May2018\n(05)','Jun2018\n(06)','Jul2018\n(07)','Aug2018\n(08)',
          'Sep2018\n(09)','Oct2018\n(10)','Nov2018\n(11)','Dec2018\n(12)',
          'Jan2019\n(13)','Feb2019\n(14)','Mar2019\n(15)','Apr2019\n(16)']

y_labels=['Jan2018 (01)','Feb2018 (02)','Mar2018 (03)',
          'Apr2018 (04)','May2018 (05)','Jun2018 (06)',
          'Jul2018 (07)','Aug2018 (08)','Sep2018 (09)',
          'Oct2018 (10)','Nov2018 (11)','Dec2018 (12)',
          'Jan2019 (13)','Feb2019 (14)','Mar2019 (15)',
          'Apr2019 (16)','May2019 (17)','Jun2019 (18)',
          'Jul2019 (19)','Aug2019 (20)','Sep2019 (21)',
          'Oct2019 (22)','Nov2019 (23)','Dec2019 (24)',
          'Jan2020 (25)','Feb2020 (26)','',
          'regular pay' ,'additional'  ,'interest']

c_labels=['0','10K','20K','30K','40K','50K','60K','70K']

grid_kws = {"height_ratios": (.9, .03), "hspace": .15}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws, figsize=(30,20))
ax = sns.heatmap(data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"},
                 cmap='Spectral_r',
                 annot=True, fmt='.0f', annot_kws={"size": 20}
                )
ax.set_xticklabels(x_labels, fontsize=20, rotation='horizontal')
ax.set_yticklabels(y_labels, fontsize=20, rotation='horizontal')
plt.style.use("dark_background")

cbar_ax.set_xticklabels(c_labels, fontsize=20);
```

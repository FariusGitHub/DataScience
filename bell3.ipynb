```python
import re
import urllib.request    
import pandas as pd
import requests
import bs4
from tqdm import tqdm

import matplotlib.pyplot as plt
import matplotlib.image as img 


urllib.request.urlretrieve("https://www.bell.ca/Mobility/Smartphones_and_mobile_internet_devices", "test.txt")

with open('test.txt', 'r') as f2:
    data = f2.read()
    soup = bs4.BeautifulSoup(data,"lxml")

#EXTRACTING THE PRICES
plist = pd.DataFrame(soup.select(".dl-tile-full-price"))          
        
#EXTRACTING DESCRIPTION AND IMAGE URLS
dgdic={postpaid.get('alt') : "https://www.bell.ca/" + \
       str(postpaid.get('data-src')) for postpaid in soup \
       .select('.dl-tile-img') if 'Prepaid' not in str(postpaid) }


# MATPLOTLIB, TQDM, LAMBDA FUNC. for data grouping, visualisation        
for j, summary in enumerate(tqdm(plist[1]
      .apply(lambda x: x.text.strip('\r\n ')))):
    im=img.imread(list(dgdic.values())[j])
    plt.imshow(im)
    plt.axis('off')
    plt.show()        
    print(list(dgdic.keys())[j])
    print(summary)
```

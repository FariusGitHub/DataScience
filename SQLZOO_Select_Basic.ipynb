```python
# SQL PRACTICE WITH PYTHON

pip install pandasql
from pandasql import sqldf 
import pandas as pd
world = pd.read_csv('https://sqlzoo.s3.us-east-2.amazonaws.com/world.csv')
pysqldf = lambda x: sqldf(x, globals())

#INSPIRED FROM https://sqlzoo.net/wiki/SELECT_basics
```

```python
#1. The example uses a WHERE clause to show the population of 'France'. 

pysqldf('''

SELECT population FROM world where name = 'Germany'

        ''')
```

```python
# Show the name and the population for 'Sweden', 'Norway' and 'Denmark'.
pysqldf('''

SELECT population FROM world where name in ('Sweden', 'Norway', 'Denmark')

        ''')
```

```python
#shows countries with an area of 250,000-300,000 sq. km. Modify it to show the
#country and the area for countries with an area between 200,000 and 250,000.

pysqldf('''

SELECT name, area FROM world
  WHERE area BETWEEN 200000 AND 250000

        ''')
```

```python
#QUIZ1
pysqldf('''

SELECT name, population 
FROM world
  WHERE population BETWEEN 1000000 AND 1270000

        ''')
```

```python
#QUIZ2
pysqldf('''

SELECT name, population 
FROM world
  WHERE name LIKE 'Al%'

        ''')
```

```python
#QUIZ3 : Select the code which shows the countries that end in A or L
pysqldf('''

SELECT count(name) 
FROM world
  WHERE name LIKE '%a' or name LIKE '%l'

        ''')

#  WHERE name LIKE '%a' or'%l' --> is not correct
```

```python
#QUIZ4
pysqldf('''

SELECT name, length(name) 
FROM world
  WHERE length(name) = 5 and continent = 'Europe'

        ''')
```

```python
#Quiz 5
pysqldf('''

SELECT  name, 
        area*2 
FROM world
  WHERE population = 76098

        ''')
```

```python
#Quiz 6 show the countries with an area larger than 50000 and a population smaller than 10000000
pysqldf('''

SELECT  name, 
        area,
        population 
FROM world
WHERE area >50000 and population < 10000000

        ''')
```

```python
#Quiz 7 shows the population density of China, Australia, Nigeria and France
pysqldf('''

SELECT  name, 
        population/area 
FROM world
WHERE name in ('China', 'Australia', 'Nigeria', 'France')

        ''')
```

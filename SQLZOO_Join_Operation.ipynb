```python
# SQL PRACTICE WITH PYTHON

from pandasql import sqldf 
import pandas as pd
game = pd.read_csv('https://sqlzoo.s3.us-east-2.amazonaws.com/game.csv')
eteam = pd.read_csv('https://sqlzoo.s3.us-east-2.amazonaws.com/eteam.csv')
goal = pd.read_csv('https://sqlzoo.s3.us-east-2.amazonaws.com/goal.csv')
pysqldf = lambda x: sqldf(x, globals())

#INSPIRED FROM https://sqlzoo.net/wiki/The_JOIN_operation
```

```python
# 1 Modify it to show the matchid and player name for all goals scored by Germany. 
#   To identify German players, check for: teamid = 'GER'

pysqldf('''

SELECT  matchid, player
from goal
where teamid = 'GER'


        ''')
```

```python
# 2 Show id, stadium, team1, team2 for just game 1012

pysqldf('''

SELECT id,stadium,team1,team2
  FROM game
where id = 1012


        ''')
```

```python
# 3 Modify it to show the player, teamid, stadium and mdate for every German goal.
pysqldf('''

SELECT player, teamid, stadium, mdate
FROM game 
JOIN goal ON (game.id=goal.matchid)
where teamid = 'GER'

        ''')
```

```python
# 4 Show the team1, team2 and player for every goal scored by a player
#   called Mario player LIKE 'Mario%'

pysqldf('''

select team1, team2, player
from game
join goal
ON (id=matchid)
where player LIKE 'Mario%'

        ''')
```

```python
# 5 Show player, teamid, coach, gtime for all goals scored in the first 10
#   minutes gtime<=10

pysqldf('''

SELECT player, teamid, coach, gtime
FROM goal 
join eteam
on (eteam.id=goal.teamid)
WHERE gtime<=10

        ''')
```

```python
# 6 List the dates of the matches and the name of the team in which 
#   'Fernando Santos' was the team1 coach.

pysqldf('''

select *
from game
join eteam
on (game.team1=eteam.id)
where coach = 'Fernando Santos'

        ''')

# select *
# from game
# join goal
# on (game.team1=goal.teamid)
# join eteam
# on (goal.teamid=eteam.id)
# where coach = 'Fernando Santos'
# above code quadruple the result as matchid differentiate each rows
```

```python
# 7 List the player for every goal scored in a game where
#   the stadium was 'National Stadium, Warsaw'

pysqldf('''

select player
from goal
join game
on (goal.matchid=game.id)
where stadium = 'National Stadium, Warsaw'
order by player ASC

        ''')

# tricky question, just list the player, no need to select player, gtime
# no need to use select distinct player, just select player --> Mario  appear 2x
```

```python
# 8 Instead show the name of all players who scored a goal against Germany.

pysqldf('''

SELECT distinct player
FROM goal
JOIN game
ON matchid=id
    WHERE teamid <> 'GER' AND (team1='GER' or team2='GER')
order by player ASC

        ''')
```

```python
# 9 Show teamname and the total number of goals scored.

pysqldf('''

SELECT teamname, count(gtime) as 'goals scored'
FROM eteam
JOIN goal
ON eteam.id=goal.teamid
group by teamname
        ''')
```

```python
# 10 Show the stadium and the number of goals scored in each stadium.

pysqldf('''

SELECT stadium, count(gtime) as 'goals scored'
FROM game
JOIN goal
ON game.id=goal.matchid
group by stadium

        ''')
```

```python
# 11 For every match involving 'POL', show the matchid, date and
#    the number of goals scored.

pysqldf('''

SELECT matchid, mdate as date, count(gtime) as 'goals scored'
FROM goal
JOIN game
ON goal.matchid=game.id
where (team1='POL' or team2='POL')
group by goal.matchid

        ''')
```

```python
# 12 For every match where 'GER' scored, show matchid, match date and the
#    number of goals scored by 'GER'

pysqldf('''

SELECT matchid, mdate as date, count(gtime) as 'goals scored'
FROM goal
JOIN game
ON goal.matchid=game.id
where teamid='GER'
group by matchid, mdate

        ''')
```

```python
# 13 List every match with the goals scored by each team as shown. 
#    This will use "CASE WHEN" which has not been explained in any
#    previous exercises.
#    Notice in the query given every goal is listed. 
#    If it was a team1 goal then a 1 appears in score1, otherwise there is a 0.
#    You could SUM this column to get a count of the goals scored by team1. 
#    Sort your result by mdate, matchid, team1 and team2.

pysqldf('''

SELECT mdate,
       team1,
       SUM(case when teamid=team1 then 1 ELSE 0 END) AS score1,
       team2,
       SUM(case when teamid=team2 then 1 ELSE 0 END) AS score2
FROM game 
LEFT JOIN goal 
ON(game.id=goal.matchid)
GROUP BY mdate, matchid, team1,team2
ORDER BY mdate, matchid, team1,team2

        ''')

# matchid is not shown, would show NaN values --> games without goals/matchid
# a good example of LEFT JOIN (gave 31 rows) and JOIN (gave 29 rows)
```

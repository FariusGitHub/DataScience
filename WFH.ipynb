```python
!pip install kora -q
from kora.selenium import wd
from selenium.webdriver.common.by import By
import time

prefix='/html/body/main/div/div/div/div[3]/div[2]/div'; 

wd.get("https://www.coursecompare.ca/courses/calgary/weclouddata/sql/")
wd.find_element(By.XPATH, "/html/body/header/div[2]/div/div[2]/div[2]/"\
                "div/div/div").click(); time.sleep(1)
display(wd)

more = int(wd.find_element(By.XPATH, "/html/body/header/div[2]/div/"\
                               "div[2]/div[2]/div/div/a").text.split(' ')[1])
load = int(more/2)

for i in range(load):
  wd.find_element(By.XPATH, prefix + "[4]/button").click(); time.sleep(1)

for i in range(more):     
    try: 
        if wd.find_element(By.XPATH,prefix + \
                  '['+str(i+2)+']/div[1]/div[1]/div[2]/div[1]/p/span')\
                  .text == 'Sales Engineer':
            wd.find_element(By.XPATH, prefix +\
                            "["+str(i+2)+"]/div[2]/button").click(); 
            wd.find_element(By.XPATH, prefix +\
                            "["+str(i+1)+"]/div[2]/div[1]/p").click();
            break
    except:
        continue

display(wd)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
import joblib

data=pd.read_csv('https://weclouddata.s3.amazonaws.com/'
     'datasets/bootcamp/final_ml_challenge_fraud/sample.csv',sep=',')

def f(X):

    return pd.concat([#TRANSFORMING x5 feature
                   X[['x5']].fillna('0.0%').apply(lambda x:\
                   (x.str.strip('%').astype(float)*100).astype(int)),
        
                   #TRANSFORMING x6 feature
                   X[['x6']].apply(lambda x: x.map({'Monday': 1,'Tuesday':\
                   2,'Wednesday': 3,'Thurday': 4,'Friday': 5,np.nan: 0})),
        
                   #TRANSFORMING x20 feature
                   X[['x20']].apply(lambda x: x.map({'Jan': 1,'Feb': 2,'Mar':\
                   3,'Apr': 4,'May': 5, np.nan: 0,'Jun': 6,'Jul': 7,'Aug':\
                   8,'Sept': 9,'Oct': 10,'Nov': 11,'Dec': 12})),
        
                   #TRANSFORMING x27 feature                   
                   X[['x27']].apply(lambda x: x.map({'Morning':1, 'Afternoon':\
                   2, 'Evening':3, 'Night':4, np.nan:0})),

                   #TRANSFORMING x49 feature                   
                   X[['x49']].apply(lambda x: x.map({True: 1,False: -1,np.nan: 0})),
    
                   #TRANSFORMING x57 feature
                   X[['x57']].fillna('$0.0').apply(lambda x:x.str.strip('$').astype(float)),
        
                   #COMBINING WITH 94 other features
                   X.drop(['x5','x6','x20','x27','x49','x57'], axis=1).fillna(0)
        
            ], axis=1)
    

pipe1 = Pipeline([('testing', FunctionTransformer(f))])
joblib.dump(pipe1, "pipe.joblib")
pipe2 = joblib.load("pipe.joblib")
pipe2.fit_transform(data)

```

I tried to improve my first python code for the WeCouldData using reverse engineering principles.
Assuming the color could be homogenous on diagonal direction as below I simply took the left diagonal as a combination of RGB scale of 0.50, 0.25, 0.25 and right diagonal as an RGB scale of 0.5, 0.75, 1 

The next step would be to identify curve formulas which are mainly in circle and parabolic equations.
Color gradation mainly happens at curve 5, 9, 10 while I simplified curve 4 to have constant color.

```python
import numpy as np; 
import matplotlib.lines as mlines; 
import matplotlib.pyplot as plt; 

fig = plt.figure(); 
plt.xlim(-6, 6), 
plt.ylim(-5, 3);

# Curve 1
x = np.linspace(-1.865, -5.529, 50) 
plt.plot(x, (4-(x+3.529)**2)**0.5, color=(0.50, 0.25, 0.50), linewidth=10);            

# Curve 2
x = np.linspace(-5.4, -5.529, 50) 
plt.plot(x, -(4-(x+3.529)**2)**0.5, color=(0.50, 0.25, 0.50), linewidth=10);           

# Curve 3 
x = np.linspace(-5.4, -3.28, 50)
plt.plot(x, 0.48*(x+4.422)**2-1.141, color=(0.50, 0.25, 0.50), linewidth=10);          

# Curve 4 (simplified)
x = np.linspace(-1.627, -3.611, 50)
plt.plot(x, 0.48*(x+2.769)**2-3.056, color=(0.50, 0.50, 0.75), linewidth=10);           

# Curve 5     
for i in np.arange(-2.719, 0, 0.2719):     
    plt.plot([i,(i+0.2719)],[0.29*(i+1.361)**2-4.181,0.29*((i+0.2719)+1.361)**2-4.181], 
             color=(0.5,0.0919*i+0.75,0.0919*i+1), linewidth=10)

# Curve 6
x = np.linspace(0, 2.719, 50)
plt.plot(x, 0.29*(x-1.141)**2-4.181, color=(0.50, 0.75, 1.00), linewidth=10);         

# Curve 7
x = np.linspace(4.943, 5.529, 50)
plt.plot(x, -(4-(x-3.529)**2)**0.5, color=(0.50, 0.75, 1.00), linewidth=10);         

# Curve 8
x = np.linspace(3.529, 5.504, 50)
plt.plot(x, -0.48*(x-4.659)**2+0.615, color=(0.50, 0.75, 1.00), linewidth=10);       

# Curve 9    
for i in np.arange(2.229, 5.529, 0.066): 
    plt.plot([i,(i+0.066)],[(4-((i)-3.529)**2)**0.5,(4-((i+0.066)-3.529)**2)**0.5],
             color=(0.5,i*0.0758+0.3311,i*0.0757+0.5811), linewidth=10)

# Curve 10
for i in np.arange(-1.3, 1.65-0.295, 0.295): 
    plt.plot([i,(i+0.295)],[(4-(i)**2)**0.5,(4-(i+0.295)**2)**0.5],
            color=(0.5,i*0.0847+0.3602,i*0.0847+0.6102), linewidth=10)

#left straight line    
for i in np.arange(-5.400, -2.719-0.2681, 0.2681): 
    plt.plot([i,(i+0.2681)],[-1.0992*i-6.6427,-1.0992*(i+0.2681)-6.6427], 
             color=(0.5,i*0.0932+0.75,i*0.0932+1.00), linewidth=10)

#top left straight
plt.plot([-1.313,-3.280], [1.505,-0.515], color=(0.50, 0.25, 0.50), linewidth=10) 

# middle straight
plt.plot([2.229,-1.627], [1.531,-2.431], color=(0.50, 0.50, 0.75), linewidth=10) 

# second right straight line
plt.plot([3.529,-0.18], [0.042,-3.827], color=(0.50, 0.75, 1.00), linewidth=10) 

# bottom right straight line
plt.plot([5.002,2.538], [-1.353,-3.648], color=(0.50, 0.75, 1.00), linewidth=10) 
```

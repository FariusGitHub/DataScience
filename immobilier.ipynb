```python
!pip install kora -q
from kora.selenium import wd
from selenium.webdriver.common.by import By
import time

wd.get("https://careerkarma.com/schools/weclouddata/")
display(wd)
prefix = '/html/body/div/div/div/div[2]/div/div/div/div/div/div'

wd.find_element(By.XPATH, prefix+"[1]/div[2]/div/div/div/a[3]").click();
time.sleep(1);

reviews = int(wd.find_element(By.XPATH, prefix + "[1]/div[2]/div/table/div[1]/"\
          "div[1]/div[2]/div/div[2]/a[1]/span").text.split(" ")[0].strip("("))
me = reviews - 38

for i in range((reviews-10)//5+(1 if (reviews-10)%5 != 0 else 0)) :
  #click LOAD MORE
  time.sleep(1); 
  wd.find_element(By.XPATH, prefix+"[2]/div[1]/div[1]/div/div/div/"\
                  "div[2]/button").click();
  #close SIGN UP pop up
  time.sleep(1)
  wd.find_element(By.XPATH, "/html/body/div[3]/div/div[2]/div/div[2]"\
                  "/button").click()

wd.find_element(By.XPATH, prefix + "[2]/div[1]/div[1]/div/div/div/div[2]/"\
                "div["+str(me)+"]/div[3]/div[2]/div[2]/label").click();
wd.find_element(By.XPATH, prefix + "[2]/div[1]/div[1]/div/div/div/div[2]/"\
                "div["+str(me)+"]/div[1]/div/h2").click();

time.sleep(1)
display(wd)
```

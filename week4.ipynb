```python
import pandas as pd
from IPython.core.display import HTML

#Related data above from Excel below to clean.
df_Surya = pd.read_excel(\
  'https://farius.s3.us-east-2.amazonaws.com/Surya+-+CAD+-+Raw+Data.xlsx'\
  ,header=0)

#Setup Empty DataFrame according to the template
df_Farius = pd.DataFrame(columns = \
  ['Supplier', 'SKU', 'Supplier Product ID', 'Product Name', 'Original Name', \
   'Description','Category Name', 'Subcategory Name', 'Height', 'Width', \
   'Depth', 'cost_usd', 'USA Price', 'cost_cad', 'CAD Price', 'Color', \
   'Technical Description', 'Default_Image', 'Image2', 'Image3', 'Image4', \
   'Image5', 'Image6', 'Style', 'Net Weight', 'UPC', 'Number of Cartons', \
   'Carton 1 Height', 'Carton 1 Width', 'Carton 1 Length', \
   'Carton 1 Volume (Cubic Feet)', 'Carton 1 Weight', 'Carton 2 Height',\
   'Carton 2 Width', 'Carton 2 Length', 'Carton 2 Volume (Cubic Feet)', \
   'Carton 2 Weight', 'Carton 3 Height', 'Carton 3 Width', 'Carton 3 Length', \
   'Carton 3 Volume (Cubic Feet)', 'Carton 3 Weight', 'Shipping Method', \
   'Country of Origin', '[any other relevant columns]'])

#Trimming the rows to clean that between 18000 and 21000 only
df_Parry=df_Surya.iloc[18000:21000,:]

#Remapping given table into the template column sequence and naming 
df_Farius['SKU'] = df_Parry['SKU']                                       #1
df_Farius['Supplier Product ID'] = df_Parry['Alias SKU']                 #2
df_Farius['Product Name'] = df_Parry['Product Description']              #3
df_Farius['Original Name'] = df_Parry['Design']                          #4
df_Farius['Description'] = df_Parry['Generic Material(s)']+', '+ \
                            df_Parry['Color(s)']+', '+ df_Parry['Style'] #5
df_Farius['Category Name'] = 'Accents'                                   #6
df_Farius['Subcategory Name'] = df_Parry['Product Type']                 #7
df_Farius['Height'] = df_Parry['Length']                                 #8
df_Farius['Width'] = df_Parry['Width']                                   #9
df_Farius['Depth'] = df_Parry['Height']                                  #10
df_Farius['cost_usd'] = df_Parry['ECOM FOB Surya']                       #11
df_Farius['USA Price'] = df_Parry['EMAP US']                             #12
df_Farius['cost_cad'] = df_Parry['ECOM CDN LANDED ']                     #13
df_Farius['CAD Price'] = df_Parry['EMAP CDN']                            #14
df_Farius['Color'] = df_Parry['Color(s)']                                #15
df_Farius['Technical Description'] = df_Parry['Material(s)'] + ', ' + \
                                      df_Parry['Shape']                  #16
df_Farius['Default_Image'] = df_Parry['Flatshot URL']                    #17
df_Farius['Image2'] = df_Parry['Roomscene URL']                          #18
df_Farius['Image3'] = df_Parry['Front URL']                              #19
df_Farius['Image4'] = df_Parry['Texture URL']                            #20
df_Farius['Image5'] = df_Parry['Fold URL']                               #21
df_Farius['Image6'] = df_Parry['Corner URL']                             #22
df_Farius['Style'] = df_Parry['Style']                                   #23
df_Farius['Net Weight'] = df_Parry['Item Weight']                        #24
df_Farius['UPC'] = df_Parry['UPC'].astype('int64')                       #25
df_Farius['Number of Cartons'] = 1                                       #26
df_Farius['Carton 1 Height'] = df_Parry['Ship Height']                   #27
df_Farius['Carton 1 Width'] = df_Parry['Ship Width']                     #28
df_Farius['Carton 1 Length'] = df_Parry['Ship Length']                   #29
df_Farius['Carton 1 Volume (Cubic Feet)'] = df_Parry['Ship Height']/12*\
                                            df_Parry['Ship Width']/12*\
                                            df_Parry['Ship Length']/12   #30
df_Farius['Carton 1 Weight'] = df_Parry['Ship Weight']                   #31
df_Farius['Carton 2 Height'] = 0                                         #32
df_Farius['Carton 2 Width'] = 0                                          #33
df_Farius['Carton 2 Length'] = 0                                         #34
df_Farius['Carton 2 Volume (Cubic Feet)'] = 0                            #35
df_Farius['Carton 2 Weight'] = 0                                         #36
df_Farius['Carton 3 Height'] = 0                                         #37
df_Farius['Carton 3 Width'] = 0                                          #38
df_Farius['Carton 3 Length'] = 0                                         #39
df_Farius['Carton 3 Volume (Cubic Feet)'] = 0                            #40
df_Farius['Carton 3 Weight'] = 0                                         #41
df_Farius['Shipping Method'] = df_Parry['Surya Ship Method']+' by Surya, ' \
                             + df_Parry['3rd Party Ship Method']\
                             + ' by 3rd party'                           #42
df_Farius['Country of Origin'] = df_Parry['Country of Origin']           #43
df_Farius['Supplier'] = 'Surya'                                          #0
# df_Farius['[any other relevant columns]'] = ''                         #44

pd.pandas.set_option('display.max_columns', None)

def path_to_image_html(path):
    return '<img src="'+ path + '" width="30" >'

#Set escape=True to change the picture back to hyperlinks
HTML(df_Farius.iloc[100:102,0:].to_html(escape=False ,formatters=dict( \
      Default_Image=path_to_image_html, Image2=path_to_image_html, \
      Image3=path_to_image_html, Image4=path_to_image_html, \
      Image5=path_to_image_html, Image6=path_to_image_html)))
```

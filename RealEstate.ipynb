```python
!pip install kora -q
from kora.selenium import wd
from selenium.webdriver.common.by import By
import time

wd.get("https://careerkarma.com/schools/weclouddata/")
display(wd)
prefix = '/html/body/div/div/div/div[2]/div/div/div/div/div/div'

#click review              
wd.find_element(By.XPATH, prefix+"[1]/div[2]/div/div/div/a[3]").click();

me = int(wd.find_element(By.XPATH, prefix+"[1]/div[2]/div/table/div[1]/div[1]"\
            "/div[2]/div/div[2]/a[1]/span").text.split(" ")[0].strip("("))-38

# click LOAD MORE 10X for example, my story must be visible in the page
for i in range(10) :
  #click LOAD MORE
  time.sleep(1); 
  wd.find_element(By.XPATH, prefix+"[2]/div[1]/div[1]/div/div/div/"\
                  "div[2]/button").click();
  #close SIGN UP pop up
  time.sleep(1)
  wd.find_element(By.XPATH, "/html/body/div[2]/div/div[2]/div/div[2]"\
                  "/button").click()

# show my story
wd.find_element(By.XPATH, prefix + "[2]/div[1]/div[1]/div/div/div/div[2]/"\
                "div["+str(me)+"]/div[3]/div[2]/div[2]/label").click();
wd.find_element(By.XPATH, prefix + "[2]/div[1]/div[1]/div/div/div/div[2]/"\
                "div["+str(me)+"]/div[1]/div/h2").click();

time.sleep(1)
display(wd)
```
